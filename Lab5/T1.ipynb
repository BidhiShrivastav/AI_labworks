```python

import torch
import torch.nn as nn
import pandas as pd

from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_data1 = pd.read_csv('assignment-data .csv')
df_data2 = pd.read_csv('assignment-data2 .csv')
df_lin_reg = pd.read_csv('linear-regression-data1.csv')
```

```python

X1 = torch.tensor(df_data1['x'].values, dtype=torch.float32)
y1 = torch.tensor(df_data1['y'].values, dtype=torch.float32)

X2 = torch.tensor(df_data2['x'].values, dtype=torch.float32)
y2 = torch.tensor(df_data2['y'].values, dtype=torch.float32)

X3 = torch.tensor(df_lin_reg['x'].values, dtype=torch.float32)
y3 = torch.tensor(df_lin_reg['y'].values, dtype=torch.float32)
     
```

```python

print(f"Data-1: {X1.shape, y1.shape}")
print(f"Data-2: {X2.shape, y2.shape}")
print(f"Data-3: {X3.shape, y3.shape}")
```

```python

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)    
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.2, random_state=42)
```

```python
import torch
import torch.nn as nn

class LinearRegressionNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)

model = LinearRegressionNN()

# Example input
x = torch.tensor([[1.0], [2.0], [3.0]])
y_pred = model(x)

print(y_pred)
```

```python
class LinearRegressionNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(
            in_features=1,
            out_features=1,
            bias=True
        )

    def forward(self, x):
        return self.linear(x)
     

torch.manual_seed(42)
model_1 = LinearRegressionNN()
     

print(f"Initial Weights for model_1: {model_1.linear.weight.item()}")
print(f"Initial Bias for model_1: {model_1.linear.bias.item()}")
     
```

# Overview of PyTorch Tensor Creation Functions

i.torch.rand(shape)-Generates a tensor filled with random values between 0 and 1, uniformly distributed.

ii.torch.zeros(shape)-Creates a tensor of the given shape where all elements are 0.

iii.torch.ones(shape)-Creates a tensor where all elements are 1.

iv.torch.eye(n)-Produces an identity matrix of size n × n, where diagonal elements are 1 and others are 0.

v.torch.arange(start, end, step)-Generates values starting from start up to (but not including) end, using a fixed step size.

vi.torch.linspace(start, end, steps)-Returns a tensor of equally spaced values between start and end (both included), with a specified number of points.

```python
rand = torch.rand(3,4)      
zeros = torch.zeros(3,4)    
ones = torch.ones(3,4)      
eye = torch.eye(3)          

arange = torch.arange(start=0, end=10, step=2)
linspace = torch.linspace(start=0, end=10, steps=5)

print(f"rand: {rand}")
print(f"zeros: {zeros}")
print(f"ones: {ones}")
print(f"eye: {eye}")
print(f"arange: {arange}")
print(f"linspace: {linspace}")

```

# Tensor Arithmetic in PyTorch
+ Operator vs torch.add()

Both + and torch.add() are used for element-wise addition of tensors, but they differ slightly in flexibility and memory control.

1.  Operator-

             i.Simple and Python-friendly syntax

             ii.Performs element-wise addition

             iii.Creates a new tensor for the result

              iv.Best suited for quick and readable operations

2. torch.add() Function-

                        i.More flexible and configurable

                        ii.Also performs element-wise addition

                        iii.Supports extra features like:Output control using out,Scaling using alpha             

```python

a = torch.tensor([1, 2, 3])
b = torch.tensor([3, 4, 5])

print(a+b)
print(a-b)
print(a*b)

# adding using add()
print(torch.add(a,b))

print(torch.add(a, b, alpha=2))
```

# @ vs. * for Tensor Multiplication
When working with torch tensors:

* (Element-wise Multiplication): This operator performs element-wise multiplication (also known as Hadamard product). It requires the tensors to have compatible shapes (either the same shape or broadcastable to the same shape). Each element in the first tensor is multiplied by the corresponding element in the second tensor.

@ (Matrix Multiplication): This operator performs matrix multiplication (or dot product for 1D tensors). It has specific shape requirements: for two matrices A (m x n) and B (n x p), the result C will be (m x p). The number of columns in the first matrix must match the number of rows in the second matrix. For 1D tensors, it performs a dot product.

```python
# Element-wise multiplication (*)
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

print(f"Tensor A: {a}")
print(f"Tensor B: {b}")
print(f"Element-wise multiplication (A * B): {a * b}")

# Matrix multiplication (@)
mat_a = torch.tensor([[1, 2],
                      [3, 4]])
mat_b = torch.tensor([[5, 6],
                      [7, 8]])

print(f"\nMatrix A: {mat_a}")
print(f"Matrix B: {mat_b}")
print(f"Matrix multiplication (A @ B): {mat_a @ mat_b}")

# Example with 1D tensors (dot product)
vec_a = torch.tensor([1, 2, 3])
vec_b = torch.tensor([4, 5, 6])

print(f"\nVector A: {vec_a}")
print(f"Vector B: {vec_b}")
print(f"Dot product (Vector A @ Vector B): {vec_a @ vec_b}")
     
```

# Deeper Dive into reshape() vs. view() Memory Behavior
The crucial distinction lies in whether the new tensor shares memory with the original tensor (a view) or if it creates a completely new block of memory (a copy).

tensor.view(): This method always attempts to return a view. This means the new tensor does not allocate new memory; it just presents the existing memory in a different shape. Because it shares memory, if you change an element in the view tensor, the original tensor will also change, and vice-versa. However, view() requires the original tensor to be contiguous in memory. If the original tensor is not contiguous (e.g., after transposing), view() will fail unless you call .contiguous() first, which might create a copy to make it contiguous.

tensor.reshape(): This method is more flexible. It will first try to return a view (sharing memory) if the memory layout allows it (i.e., if the tensor is contiguous and the new shape is compatible without reordering data). If the memory layout does not allow for a view (e.g., the tensor is not contiguous, or the new shape would require reordering elements that are not adjacent in memory), reshape() will automatically return a copy of the data in new memory. Changes to a copied tensor will not affect the original, and vice-versa.

```python
import torch

t = torch.arange(10, 19, 1)  # values from 10 to 18

print(t.reshape(3, 3))
print(t.view(3, 3))
print(t)
```

# Moving to GPU in Machine Learning
Why GPUs are Preferred over CPUs-
1. CPU (Central Processing Unit)-


                                i.Designed for general-purpose computing

                                ii.Has few but powerful cores

                                iii.Best for sequential and complex tasks

                                iv.Works on tasks one after another

2. GPU (Graphics Processing Unit)-

                               i.Designed for parallel processing

                               ii. contains thousands of smaller cores

                               iii. Best for repetitive and large-scale computations

                            iv.Performs many operations simultaneously

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

t = torch.tensor([1., 2., 3.])
t_gpu = t.to(device)
print(t_gpu.device)

# move back to CPU (needed before NumPy conversion)
t_cpu = t_gpu.cpu()
print(t_cpu.numpy())
     
```
